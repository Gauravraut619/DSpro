2. How does the survival rate vary based on passenger class (Pclass)?

```python
# Calculate survival rate by passenger class
survival_by_pclass =df.groupby('Pclass')['Survived'].mean()



```

```python
# Plot survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_by_pclass.index, y=survival_by_pclass.values, palette="viridis")
plt.title('Survival Rate by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 1)
plt.show()
```

Ans:- This suggests that higher-class passengers had better chances of survival.

3. Is there a relationship between the fare paid and the likelihood of survival?

```python
# Create a scatter plot to examine the relationship between Fare and Survival
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Survived'], y=df['Fare'], palette="coolwarm")
plt.title('Relationship Between Fare and Survival', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Fare Paid', fontsize=12)
plt.xticks([0, 1], labels=['No', 'Yes'], fontsize=10)
plt.yticks(fontsize=10)
plt.yscale('log')
# Use log scale for better visualization due to outliers
plt.show()
```

Ans:- Higher Fares: Passengers who survived generally paid higher fares. The median fare for survivors is notably higher than for those who did not survive.
Lower Fares: Passengers who paid lower fares were less likely to survive.
This suggests that passengers in more expensive classes (linked with higher fares) had better chances of survival.

4. How does the survival rate differ between male and female passengers?


```python
# Calculate survival rate by gender
survival_by_gender = df.groupby('Sex')['Survived'].mean()

```

```python
# Plot survival rate by gender
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values, palette="coolwarm")
plt.title('Survival Rate by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 1)
plt.show()
```

Females: Had a significantly higher survival rate compared to males.
Males: Had a much lower survival rate.
This indicates that gender played a crucial role in survival, likely influenced by the "women and children first" policy during the evacuation

What impact does the number of siblings/spouses (SibSp) or parents/children
(Parch) aboard have on survival?

```python
# Calculate survival rates based on SibSp
sibsp_survival = df.groupby('SibSp')['Survived'].mean()
```

```python
# Calculate survival rates based on Parch
parch_survival = df.groupby('Parch')['Survived'].mean()
```

```python
# Plot survival rates for SibSp
plt.figure(figsize=(12, 6))
plt.bar(sibsp_survival.index, sibsp_survival.values, alpha=0.7, color='blue')
plt.title('Survival Rate by Number of Siblings/Spouses Aboard (SibSp)', fontsize=14)
plt.xlabel('Number of Siblings/Spouses Aboard (SibSp)', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(sibsp_survival.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
# Plot survival rates for Parch
plt.figure(figsize=(12, 6))
plt.bar(parch_survival.index, parch_survival.values, alpha=0.7, color='green')
plt.title('Survival Rate by Number of Parents/Children Aboard (Parch)', fontsize=14)
plt.xlabel('Number of Parents/Children Aboard (Parch)', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(parch_survival.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The bar charts display the survival rates based on the number of siblings/spouses (SibSp) and parents/children (Parch) aboard:

SibSp:

Survival rates vary depending on the number of siblings or spouses aboard.
Passengers with 1 sibling/spouse aboard have a higher survival rate.
Passengers with no siblings/spouses or many (e.g., 3 or more) have lower survival rates.
Parch:

Passengers with 1 or 2 parents/children aboard have higher survival rates.
Passengers with no parents/children or a larger family group (e.g., 3 or more) tend to have lower survival rates.
This suggests that being accompanied by a small family group might have improved the chances of survival. Large family groups might have faced more difficulties during evacuation.

Are there any significant differences in survival rates among passengers embarked
from different ports?

```python
# Calculate survival rates based on embarkation ports
embarkation_survival = df.groupby('Embarked')['Survived'].mean()

# Plot survival rates by embarkation port
plt.figure(figsize=(8, 5))
plt.bar(embarkation_survival.index, embarkation_survival.values, color=['red', 'blue', 'green'], alpha=0.7)
plt.title('Survival Rate by Embarkation Port', fontsize=14)
plt.xlabel('Embarkation Port', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(embarkation_survival.index, labels=['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Display survival rates numerically
embarkation_survival
```

The survival rates by embarkation port are as follows:

Cherbourg (C): 55.36%
Queenstown (Q): 38.96%
Southampton (S): 33.70%
Key Insights:
Passengers who embarked from Cherbourg had significantly higher survival rates compared to those from Queenstown and Southampton.
Southampton, the most common embarkation point, had the lowest survival rate.
This disparity could be influenced by factors such as the socio-economic status of passengers boarding at different ports or the distribution of cabin classes.
