```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Iris.csv')
```

What is the distribution of sepal lengths and widths among the Iris species?

```python
# Plotting the distribution of sepal lengths and widths by species
plt.figure(figsize=(12, 6))

# Plot for Sepal Length by Species
plt.subplot(1, 2, 1)
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.hist(subset['SepalLengthCm'], bins=10, alpha=0.6, label=species)
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()

# Plot for Sepal Width by Species
plt.subplot(1, 2, 2)
for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.hist(subset['SepalWidthCm'], bins=10, alpha=0.6, label=species)
plt.title('Distribution of Sepal Width by Species')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

Here are the distributions of sepal lengths and widths among the Iris species:

Sepal Length Distribution:

Iris-setosa has a shorter sepal length range compared to the other species.
Iris-versicolor and Iris-virginica have overlapping ranges but show slightly longer sepal lengths overall.
Sepal Width Distribution:

Iris-setosa generally has wider sepals, showing a distinct separation from the other species.
Iris-versicolor and Iris-virginica have more overlap in sepal width but are still distinguishable.

How does the petal length vary across different Iris species?

```python
# Boxplot to show the variation of Petal Length across different Iris species
plt.figure(figsize=(8, 6))
df.boxplot(column='PetalLengthCm', by='Species', grid=False, notch=True, patch_artist=True)

plt.title('Variation of Petal Length Across Iris Species')
plt.suptitle('')  # Remove the automatic boxplot title
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

The boxplot above shows the variation in petal length across the three

Iris species:


Iris-setosa:

Petal lengths are relatively small and consistent, with very little variation.

Iris-versicolor:

Petal lengths are longer compared to Iris-setosa and have moderate variation.

Iris-virginica:

Petal lengths are the longest among the species and show the widest variation.

This indicates that petal length is a strong feature for distinguishing between Iris species.

Is there a relationship between sepal length and petal length?

```python
# Scatter plot to examine the relationship between sepal length and petal length
plt.figure(figsize=(8, 6))

for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.scatter(subset['SepalLengthCm'], subset['PetalLengthCm'], label=species, alpha=0.7)

plt.title('Relationship Between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

The scatter plot shows a clear relationship between sepal length and petal length across the Iris species:

Iris-setosa:

Petal lengths and sepal lengths are relatively small and clustered, indicating a weaker relationship.

Iris-versicolor:

A moderate positive correlation is evident between sepal length and petal length.

Iris-virginica:

A strong positive correlation is observed, with both sepal and petal lengths being larger on average.

This indicates that sepal length and petal length are positively correlated, with the strength of the relationship varying among the species.

How does the petal width vary across different Iris species?

```python
# Boxplot to show the variation of Petal Width across different Iris species
plt.figure(figsize=(8, 6))
df.boxplot(column='PetalWidthCm', by='Species', grid=False, notch=True, patch_artist=True)

plt.title('Variation of Petal Width Across Iris Species')
plt.suptitle('')  # Remove the automatic boxplot title
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

The boxplot above highlights the variation in petal width across the three Iris species:

Iris-setosa:

Petal widths are the smallest and most consistent among the species.

Iris-versicolor:

Petal widths are moderate in size with slightly higher variability compared to Iris-setosa.

Iris-virginica:

Petal widths are the largest, showing the widest variation.

This demonstrates that petal width, like petal length, is a strong distinguishing feature among the species.

Are there any patterns observable in the pairwise relationships between features
(sepal length, sepal width, petal length, petal width)?

```python
# Ensure all numerical columns are correctly formatted as numeric
numeric_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values in numerical columns
df = df.dropna(subset=numeric_columns)

# Retry the pairplot visualization
sns.pairplot(df, hue='Species', diag_kind='hist', corner=True, height=2.5)

plt.suptitle('Pairwise Relationships Between Features', y=1.02)
plt.show()
```

The pairplot shows the pairwise relationships between the features (sepal length, sepal width, petal length, petal width) across the Iris species:

Diagonal: Displays the distribution of each feature for the three species using histograms.

Off-Diagonal: Displays scatter plots showing relationships between pairs of features, colored by species.

Observations:
Petal Length vs. Petal Width:
A strong positive correlation is evident, especially for Iris-versicolor and Iris-virginica.

Sepal Length vs. Petal Length:
Another positive correlation is clear, with distinct clusters for each species.

Sepal Width:
Shows weaker relationships with other features and more overlap between species.

Can we distinguish between Iris species based on sepal and petal measurements?

```python
# Scatter plot: Petal Length vs Petal Width to show distinction among Iris species
plt.figure(figsize=(8, 6))

for species in df['Species'].unique():
    subset = df[df['Species'] == species]
    plt.scatter(subset['PetalLengthCm'], subset['PetalWidthCm'], label=species, alpha=0.7)

plt.title('Distinction Between Iris Species Based on Petal Measurements')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

The scatter plot above illustrates the distinction between Iris species based on petal measurements:

Iris-setosa:

Forms a distinct cluster with smaller petal lengths and widths, clearly separable from the other species.

Iris-versicolor and Iris-virginica:

Show some overlap, but the trend of increasing petal length and width helps distinguish them.
This plot highlights why petal dimensions are highly effective for species classification
